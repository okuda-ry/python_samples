## 売り上げを補正する

```python
sales["item_name"] = sales["item_name"].str.upper()
sales["item_name"] = sales["item_name"].str.replace("　","")
sales["item_name"] = sales["item_name"].str.replace(" ","")
sales.sort_values(by="item_name",ascending=True)
```

```python
pd.unique(sales["item_name"])
```

```python
flg_is_null = sales["item_price"].isnull()
for flg in list(sales.loc[flg_is_null,"item_name"].unique()):
    price = sales.loc[(~flg_is_null) & (sales["item_name"]==flg),"item_price"].max()
    sales.loc[(flg_is_null) & (sales["item_name"]==flg),"item_price"] = price
```

```python
sales
```

```python
for flg in list(sales["item_name"].unique()):
    print("=="+flg+"==")
    print("最大値："+str(sales.loc[sales["item_name"]==flg,"item_price"].max())+" 最小値："+str(sales.loc[sales["item_name"]==flg,"item_price"].min(skipna=False)))
```

## 顧客名を補正する

```python
customers["顧客名"].head()
```

```python
sales["customer_name"].head()
```

```python
customers["顧客名"] = customers["顧客名"].str.replace("　","")
customers["顧客名"] = customers["顧客名"].str.replace(" ","")
customers.head()
```

```python
customers.head(3)
```

```python
flg_is_serial = customers["登録日"].astype(str).str.isdigit()
flg_is_serial.sum()
```

```python
fromSerial = pd.to_timedelta(customers.loc[flg_is_serial,"登録日"].astype(int)-2,unit="D")+pd.Timestamp("1900-01-01")
# Excelのシリアル値は1900年1月1日を1日目としているが、実際には1900年1月0日が起点となっているため、-2日する必要がある
fromString = pd.to_datetime(customers.loc[~flg_is_serial,"登録日"])
# 文字列として認識された日付をdatetime型に変換
customers["登録日"] = pd.concat([fromSerial,fromString]).sort_index()
# 両者を結合し、元のインデックス順に並び替え
```

```python
customers["登録年月"] = customers["登録日"].dt.strftime("%Y%m")
rslt = customers.groupby("登録年月").count()["顧客名"]
rslt
```

```python
flg_is_serial = customers["登録日"].astype(str).str.isdigit()
flg_is_serial.sum()
```
